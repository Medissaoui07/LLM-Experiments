```python
def convert_to_chatML(example):
  # Extract user and assistant messages
  user_message = next((msg["content"] for msg in example["messages"] if msg["role"] == "user"), None)
  assistant_message = next((msg["content"] for msg in example["messages"] if msg["role"] == "assistant"), None)

  # Return in the desired format
  return {
      "messages": [
          {"role": "user", "content": user_message},
          {"role": "assistant", "content": assistant_message},
      ]
  }

# Apply the function
ds.map(convert_to_chatML)
```

```python
import torch
from trl import SFTTrainer, SFTConfig, setup_chat_format
from datasets import load_dataset
from transformers import AutoModelForCausalLM, AutoTokenizer

device = "cuda" if torch.cuda.is_available() else "cpu"

# Load the model and tokenizer using distinct variable names
model = AutoModelForCausalLM.from_pretrained("HuggingFaceTB/SmolLM2-135M").to(device)
tokenizer2 = AutoTokenizer.from_pretrained("HuggingFaceTB/SmolLM2-135M")
print(model)

# Check if the tokenizer has a chat_template, if not, try to add a default one
if tokenizer2.chat_template is None:
    print("Tokenizer does not have a chat_template. Attempting to add a default one.")
    try:
        tokenizer2.chat_template = "{% for message in messages %}\n{% if message['role'] == 'user' %}{{ '### User:' + message['content'] + '\\n' }}{% elif message['role'] == 'assistant' %}{{ '### Assistant:' + message['content'] + '\\n' }}{% endif %}\n{% endfor %}"
    except Exception as e:
        print(f"Error adding default chat template: {e}")

# Generate a sample from the base model
prompt = "give me a fact about Tunisia"
messages = [{"role": "user", "content": prompt}]

try:
    # Apply chat template using the tokenizer object (tokenizer2)
    prompt = tokenizer2.apply_chat_template(messages, tokenize=False)

    input_ids = tokenizer2(prompt, return_tensors="pt").to(device)
    output = model.generate(**input_ids, max_new_tokens=100)
    print(tokenizer2.decode(output[0]))

except AttributeError as e:
    print(f"Still getting an AttributeError: {e}")
    print("It seems the tokenizer might not be compatible with `apply_chat_template`.")
    print("Let's try a manual formatting approach.")

    formatted_prompt = "### User: give me a fact about Tunisia\n### Assistant:"
    input_ids = tokenizer2(formatted_prompt, return_tensors="pt").to(device)
    output = model.generate(**input_ids, max_new_tokens=100)
    print(tokenizer2.decode(output[0]))
```

```python
training_args = SFTConfig(
    output_dir="./sft_output",
    max_steps=1000,
    per_device_train_batch_size=4,
    learning_rate=5e-5,
    logging_steps=10,
    save_steps=100,
    eval_strategy="steps",
    eval_steps=50,
)

# Initialize trainer
trainer = SFTTrainer(
    model=model,
    args=training_args,
    train_dataset=ds["train"],
    eval_dataset=ds["test"],
    processing_class=tokenizer2,
)
```

```python
trainer.train()
trainer.save_model("./fine-tuned-model")
```
